```python
from telethon import TelegramClient
import pandas as pd
import nest_asyncio
import csv

nest_asyncio.apply()

api_id = '26496282'  
api_hash = 'b08c68442a0aff65dd99089e165b5816' 
phone_number = '6204916223'

channel_1 = 'https://t.me/stockmarketupdates4'
channel_2 = 'https://t.me/niftydirection'

client = TelegramClient('session_name', api_id, api_hash)

async def scrape_channel(channel_name, output_file):
    async with client:
        print(f"Scraping messages from {channel_name}...")
        channel = await client.get_entity(channel_name)
        messages = await client.get_messages(channel, limit=5000) 
        
        with open(output_file, 'w', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(['Date', 'Sender', 'Message'])  
            
            for msg in messages:
                writer.writerow([msg.date, msg.sender_id, msg.text])

async def main():
    await scrape_channel(channel_1, 'channel_1_messages.csv')
    await scrape_channel(channel_2, 'channel_2_messages.csv')
    print("Scraping completed!")

client.loop.run_until_complete(main())
```
